```python
import numpy as np
import scipy.io
import pickle
import matplotlib.pyplot as plt
%cd '/content/drive/MyDrive/EEG project'
```

```python
f=open('./swav_train/stats0.pkl','rb')
df = pickle.load(f)
df
```

```python
train_loss = df['loss']
plt.plot(train_loss, label='swav loss')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend()
plt.show()
```

```python
f1=open('./eval/stats0.pkl','rb')
df1 = pickle.load(f1)
df1.max()
```

```python
train_loss = df1['loss']
val_loss = df1['loss_val']
plt.plot(train_loss,label='train')
plt.plot(val_loss,label='validation')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend()
plt.savefig('1.jpg')
plt.show()
```

```python
train_acc = df1['prec1']
val_acc = df1['prec1_val']
plt.plot(train_acc,label='train')
plt.ylim((30,70))
plt.plot(val_acc,label='validation')
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.legend()
plt.savefig('2.jpg')
plt.show()
```
